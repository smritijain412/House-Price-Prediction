```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
from sklearn.metrics import r2_score,accuracy_score
```

```python
def compar(df_final):
    #Split data
    X=df_final.iloc[:,1:]
    Y=df_final.iloc[:,0]
    X_train,x_test,Y_train,y_test=train_test_split(X,Y,random_state=35)
    print("Shape",X_train.shape,Y_train.shape,x_test.shape,y_test.shape)
    print("*******************************************************************")
    #Use model
    LR=LinearRegression()
    LR.fit(X_train,Y_train)
    print("Model Score-->",LR.score(x_test,y_test))
    print("*******************************************************************")
    y_pred=LR.predict(x_test)
    print("r2 score",r2_score(y_test,y_pred))
    print("*******************************************************************")
    # Calculate VIF for each variable
    df_with_const = add_constant(df_final)
    v_d = pd.DataFrame()
    v_d["Variable"] = df_with_const.columns
    v_d["VIF"] = [variance_inflation_factor(df_with_const.values, i) for i in range(df_with_const.shape[1])]
    # Display the result
    print("Multi-Colinearity score",v_d)
    for x in [0.1,0.3,0.5,0.6]:
        ridge = Ridge(alpha=x)
        ridge_double=Ridge(alpha=x*2)
        ridge.fit(X_train,Y_train)
        ridge_double.fit(X_train,Y_train)
        pred_r = ridge.predict(x_test)
        pred_r2=ridge_double.predict(x_test)
        print("Alpha score-->",x,"RidgeScore-->",r2_score(y_test, pred_r),"RidgeScoredouble-->",r2_score(y_test, pred_r2))
        print("*****************************************************************************************************")
        lasso = Lasso(alpha=x)
        lasso_double=Lasso(alpha=x*2)
        lasso.fit(X_train,Y_train)
        lasso_double.fit(X_train,Y_train)
        pred_l = lasso.predict(x_test)
        pred_l2=lasso_double.predict(x_test)
        print("Alpha score-->",x,"lassoScore-->",r2_score(y_test, pred_l),"lassoScoredouble-->",r2_score(y_test, pred_l2))
```
